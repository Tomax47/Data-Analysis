```python
import pandas as pd
```

```python
df = pd.read_excel('med.xlsx', engine='openpyxl')
df
```

```python
# Removing empty columns
cleaned_ds = df.dropna(subset=['Systolic Blood Pressure'])
cleaned_ds
```

```python
# Ensuring the Systolic Blood Pressure being between 90 and 230 by filtering out the outliers
filtered_ds = cleaned_ds[~((cleaned_ds['Systolic Blood Pressure'] < 90) | (cleaned_ds['Systolic Blood Pressure'] > 230))]
filtered_ds
```

```python
# Updating non-numerical values to binaries
mappings = {
    'Hypertension': {'Yes': 0, 'No': 1},
    'Patient Gender': {'M': 0, 'F': 1},
    'Smoking Status': {'Never': 1, 'Yes': 0},
    'Primary / Secondary CVRM': {'Primary': 0, 'Secondary': 1}
}

filtered_ds = filtered_ds.replace(mappings)

filtered_ds
```

**Point-2**

```python
# Categorizing the data based on the location being an Urban "г." -> 0 /rural "ЦРБ" -> 1
def categorize_location(name):
    if "г." in name:
        return "0"
    else:
        return "1"

filtered_ds['Organisation Name (CVRM Treatment)'] = filtered_ds['Organisation Name (CVRM Treatment)'].apply(categorize_location)
filtered_ds
```

```python
# Dividing sample into urban and rural
urban_sample = filtered_ds[filtered_ds['Organisation Name (CVRM Treatment)'] == "0"]
rural_sample = filtered_ds[filtered_ds['Organisation Name (CVRM Treatment)'] == "1"]

urban_sample.head(5)
```

```python
variant_col = 'Systolic Blood Pressure'

# Urban sample statistics
urban_size = len(urban_sample)
urban_mean = urban_sample[variant_col].mean()
urban_variance = urban_sample[variant_col].var()
urban_std_dev = urban_sample[variant_col].std()

# Rural sample statistics
rural_size = len(rural_sample)
rural_mean = rural_sample[variant_col].mean()
rural_variance = rural_sample[variant_col].var()
rural_std_dev = rural_sample[variant_col].std()

print("Urban Sample Statistics:")
print(f"Sample size: {urban_size}\nMean: {urban_mean}\nVariance: {urban_variance}\nStandard Deviation: {urban_std_dev}\n")

print("Rural Sample Statistics:")
print(f"Sample size: {rural_size}\nMean: {rural_mean}\nVariance: {rural_variance}\nStandard Deviation: {rural_std_dev}")
```

Point-3

```python
# Calculating the correlation for the whole dataset
filtered_ds = filtered_ds.select_dtypes(include=[float, int]) # Filtering out non-numerical cols

correlation_matrix = filtered_ds.corr()
print(correlation_matrix)
```

```python
# Exporting & downloading data as xlsx
from google.colab import files

dataframes = [
    (filtered_ds, 'filtered_ds_med.xlsx'),
    (rural_sample, 'urban_sample_mde.xlsx'),
    (urban_sample, 'rural_sample_med.xlsx')
]

for df, filename in dataframes:
    df.to_excel(filename, index=False, engine='openpyxl')

for _, filename in dataframes:
    files.download(filename)
```
